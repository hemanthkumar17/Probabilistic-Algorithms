```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
y = [1 - (1 - x**k)**50 for k in range(1, 8)]

plt.xlabel("Jaccard Similarity (Jx)")
plt.ylabel("Retrieval Probability (Px)")
for k in range(7):
    plt.plot(x, y[k], label=f"k = {k+1}")

plt.legend()
```

```python

import matplotlib.pyplot as plt
import numpy as np

l_space = [5, 10, 20, 50, 100, 150, 200]
x = np.linspace(0, 1, 100)
y = [1 - (1 - x**4)**l for l in l_space]

plt.xlabel("Jaccard Similarity (Jx)")
plt.ylabel("Retrieval Probability (Px)")
for l in range(len(l_space)):
    plt.plot(x, y[l], label=f"l = {l_space[l]}")

plt.legend()
```
